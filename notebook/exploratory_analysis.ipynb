```python
import folium
from folium.plugins import HeatMap

# Extract list of coordinates and convert to floats
heat_data = df['Coordinates'].apply(lambda x: [float(i) for i in x.split(',')]).tolist()

# Create map
m = folium.Map(location=[sum(x[0] for x in heat_data)/len(heat_data),
                         sum(x[1] for x in heat_data)/len(heat_data)], zoom_start=5)

# Add heat map
HeatMap(heat_data).add_to(m)

m.save("heatmap.png")
```
